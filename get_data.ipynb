```python
bitcoin_df = pd.DataFrame(target_data,columns=['Date','High','Low',
                              'Mid',
                              'Last',
                              'Bid',
                              'Ask',
                              'Volume'])

bitcoin_df
    
```

```python
volume_df = bitcoin_df[["Date","Volume"]]
volume_df
```

```python
v2021_df = volume_df.loc[volume_df["Date"]>="2021-01-01"]
v2020_df = volume_df.loc[(volume_df["Date"]>="2020-01-01") & (volume_df["Date"]<="2020-12-31")]
v2019_df = volume_df.loc[(volume_df["Date"]>="2019-01-01") & (volume_df["Date"]<="2019-12-31")]
v2018_df = volume_df.loc[(volume_df["Date"]>="2018-01-01") & (volume_df["Date"]<="2018-12-31")]
v2017_df = volume_df.loc[(volume_df["Date"]>="2017-01-01") & (volume_df["Date"]<="2017-12-31")]
```

```python
avg_v2021 = v2021_df["Volume"].mean()
avg_v2020 = v2020_df["Volume"].mean()
avg_v2019 = v2019_df["Volume"].mean()
avg_v2018 = v2018_df["Volume"].mean()
avg_v2017 = v2017_df["Volume"].mean()
```

```python
avg_vol_df = pd.DataFrame({"Year":["2021", "2020", "2019", "2018", "2017"],
                           "Average_Volume":[avg_v2021, avg_v2020, avg_v2019, avg_v2018, avg_v2017]})

avg_vol_df
```

```python
avg_vol_df.to_csv("./static/data/average_volume.csv")
```
